```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


# Datos de entrenamiento
densidades_planetas = dFFinal0206byesi['dens_pl_ert']
radios_planetas = dFFinal0206byesi['pl_rade']
temperaturas_estrella = dFFinal0206byesi['st_teff']
equilibrio_temperatura = dFFinal0206byesi['pl_eqt']
Esival = dFFinal0206byesi['Esi_value']

# Preprocesamiento de datos
X = list(zip(densidades_planetas, radios_planetas, temperaturas_estrella, equilibrio_temperatura))
y = Esival

# DivisiÃ³n de datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=52)

# CreaciÃ³n y entrenamiento del modelo KNN Regressor
k = 4  # NÃºmero de vecinos a considerar
knn = KNeighborsRegressor(n_neighbors=k)
knn.fit(X_train, y_train)

# PredicciÃ³n en el conjunto de prueba
y_pred = knn.predict(X_test)

# EvaluaciÃ³n del modelo
mse = mean_squared_error(y_test, y_pred)
print(f"Error cuadrÃ¡tico medio: {mse}")

# PredicciÃ³n para un nuevo planeta
# [Densidad, Radio, Temperatura Estrella, Equilibrio de Temperatura]
nuevo_planeta = [[5.5, 3500, 5777, 250]]  # Ejemplo de nuevos valores de densidad, radio, temperatura estrella y equilibrio de temperatura
similitud = knn.predict(nuevo_planeta)
print(f"La similitud del nuevo planeta con la Tierra es: {similitud}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import learning_curve

# Datos de entrenamiento
densidades_planetas = dFFinal0206byesi['dens_pl_ert']
radios_planetas = dFFinal0206byesi['pl_rade']
temperaturas_estrella = dFFinal0206byesi['st_teff']
equilibrio_temperatura = dFFinal0206byesi['pl_eqt']
Esival = dFFinal0206byesi['Esi_value']

# Preprocesamiento de datos
X = list(zip(densidades_planetas, radios_planetas, temperaturas_estrella, equilibrio_temperatura))
y = Esival

# DivisiÃ³n de datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=52)

# CreaciÃ³n y entrenamiento del modelo KNN Regressor
k = 4  # NÃºmero de vecinos a considerar
knn = KNeighborsRegressor(n_neighbors=k)

# GeneraciÃ³n de la curva de aprendizaje
train_sizes, train_scores, test_scores = learning_curve(
    knn, X_train, y_train, cv=5, scoring='neg_mean_squared_error', 
    train_sizes=np.linspace(0.1, 1.0, 10)
)

# CÃ¡lculo del error cuadrÃ¡tico medio promedio en cada conjunto de entrenamiento
train_mse = -np.mean(train_scores, axis=1)
test_mse = -np.mean(test_scores, axis=1)

# Graficar la curva de aprendizaje
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mse, 'o-', color='r', label='Entrenamiento')
plt.plot(train_sizes, test_mse, 'o-', color='g', label='Prueba')
plt.xlabel('TamaÃ±o del conjunto de entrenamiento')
plt.ylabel('Error cuadrÃ¡tico medio')
plt.title('Curva de Aprendizaje')
plt.legend(loc='best')
plt.savefig('apr4.png')
plt.show()

# Entrenamiento final con todo el conjunto de entrenamiento
knn.fit(X_train, y_train)

# PredicciÃ³n en el conjunto de prueba
y_pred = knn.predict(X_test)

# EvaluaciÃ³n del modelo
mse = mean_squared_error(y_test, y_pred)
print(f"Error cuadrÃ¡tico medio: {mse}")

# PredicciÃ³n para un nuevo planeta
# [Densidad, Radio, Temperatura Estrella, Equilibrio de Temperatura]
nuevo_planeta = [[5, 3000, 5000, 300]]  # Ejemplo de nuevos valores de densidad, radio, temperatura estrella y equilibrio de temperatura
similitud = knn.predict(nuevo_planeta)
print(f"La similitud del nuevo planeta con la Tierra es: {similitud}")
```
